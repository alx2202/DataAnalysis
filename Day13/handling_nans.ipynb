# Handling `NaN`s

What is `NaN`? See [Pandas docs](https://pandas.pydata.org/docs/user_guide/missing_data.html)



```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/alx2202/DataAnalysis/raw/main/Day13/employees_ext.csv'
emps = pd.read_csv(url, delimiter=',', parse_dates=['hire_date'], index_col='employee_id')
emps
```

```python
emps.sample(15)
```

```python
emps.isnull()
```

```python
# Quick method to see if we have any missing values in our dataframe divided into columns.
# emps.isnull().any()
emps.isnull().any(axis=0)
```

```python
emps.isnull().any(axis=1).sample(20)
```

## Axes in Pandas

[Axes visually explained](https://stackoverflow.com/a/25774395/779084)

- axis = 0 (this is one is used by default) merges the rows and gives the results oriented into columns
- axis = 1 merges columns and gives the results oriented into rows

```python
# get all the rows that contain at least one missing value
# emps.isnull().any(axis=1)
emps[emps.isnull().any(axis=1)]
```

```python
# get all the rows that don't have any missing value
emps[~emps.isnull().any(axis=1)]
```

## How we can replace NaN values with different ones?

### `.replace()`

```python
# I'm replacing NaN throughout DataFrame, all the columns and rows are impacted
# by default the original DataFrame is not impacted (you get a copy) unless you pass `inplace=True`.
emps.replace(np.NaN, 0.0)
```

### `.fillna()`

[`.fillna()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html?highlight=fillna#pandas.DataFrame.fillna) - replaces missing values (like `NaN`, `NaT`) by given value.

```python
emps.fillna(0.0)
```

```python
emps.fillna({
    'manager_id': 0.0,
    'commission_pct': 0.99,
})
```

```python
emps.sort_values(['first_name', 'last_name'], inplace=True)
emps.head(10)
```

```python
# to the field containing NaN we will put previous non-NaN value
emps.fillna(method='ffill').head(10)
```

```python
emps.head(10)
```

```python
# to the field containing NaN we will put first non-NaN value from later rows
emps.fillna(method='bfill').head(10)
```

```python
# we can limit how many rows with NaN will be impacted by such operation
emps.fillna(method='bfill', limit=1).head(10)
```

## Why we should check and handle `NaN`s in our data?

- `NaN`s will impact many operations we will be doing and this impact may cause unexcepted results of mathematic functions.

```python
emps
```

`.mean()` - is calulated based on non-empty values, so empty/missing values are not included into the calculation. So the following calculation will take only those values from `commission_pct` column where we don't have `NaN`.

```python
emps.commission_pct.mean()
```

```python
emps[~emps.commission_pct.isnull()]
```

```python
emps[~emps.commission_pct.isnull()].commission_pct.mean()
```

```python
# if I want to calculate an average (mean) that includes all the rows
emps.commission_pct.fillna(0.0).mean()
```

```python
emps.commission_pct.mean(skipna=False)
```

```python
# emps.commission_pct.isnull()
emps.commission_pct.isna()
```

```python
#emps.commission_pct.notnull()
emps.commission_pct.notna()
```

```python
# Employess that don't take commission
emps[emps.commission_pct.isna()]
```

```python
# Employess that take commission
emps[emps.commission_pct.notna()]
```

```python
sum(emps.commission_pct.isna()), sum(emps.commission_pct.notna())
```

## Removing rows that contain `NaN`s

If we don't want to replace `NaN` by something else we can always ignore those rows that contain `NaN` for our calculations, we can remove them as well from our DataFrame. 

For that useful method we can use is [`.dropna()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html?highlight=dropna#pandas.DataFrame.dropna).

```python
emps.dropna()  # removing all rows that contain NaN in any column
```

`how`

- `any` : If any NA values are present, drop that row or column.
- `all` : If all values are NA, drop that row or column - this option is really useful when we are cleaning the data

```python
emps.dropna(how='all')
```

We can use `subset` argument to provide an information which columns should be checked for `NaN` and if they are found in this subset, then the row will be dropped.

```python
emps.sort_index(inplace=True)
emps
```

```python
emps.dropna(subset=['manager_id'])
```
