I want to see 10 matches with the biggest difference in scored goals.

- we need to read the CSV file with matches.csv.
- we need to create a separate column with the difference between the goals for each match
    - use `abs()` function to get positive results `abs(-3) -> 3`
    - we can assign the result to new column in the DataFrame

```python
import pandas as pd
```

```python
url = 'https://github.com/alx2202/DataAnalysis/raw/main/Day10/matches.csv'
matches = pd.read_csv(url, parse_dates=['date'], sep=',')
matches
```

[`abs(x)`](https://docs.python.org/3/library/functions.html?highlight=abs#abs) - abbreviation of absolute

```python
# matches['something'] = matches['sometheing'] + matches['sometheing'] 
matches['goals_diff'] = abs(matches['home_team_goal'] - matches['away_team_goal'])  # 
matches
```

`sort_values()` sorts the DataFrame by columns, by default in the ascending order.

```python
matches.sort_values('goals_diff', ascending=False)
```

```python
matches.sort_values('goals_diff', ascending=False).head(10)
```

```python
matches.sort_values('goals_diff', ascending=False).iloc[0:10]
```

```python
matches.sort_values('goals_diff', ascending=False).iloc[0:15]
```

```python
matches.nlargest(n=10, columns='goals_diff', keep='all')
```
