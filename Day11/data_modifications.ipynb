## How we can change column names?

What kind of problems we have with this data set, with those columns?
- we have spaces in the names and it will cause access to those columns a bit more complicated as I will have to use only ['column name'] approach

````python
# df.columns[0] = 'Cat'  # KO: TypeError: Index does not support mutable operations
````

Although we can't change a particular column we can change all the column names at once by assigning new list with names.

````python
new_column_names = []
for column in df.columns:
    new_column_names.append(column.lower())
df.columns = new_column_names
df
````

I'd like to do several changes to each column name:
- replace spaces with `_`
- replace `%` with `Percentage`
- replace `(` and `)` with an empty string `''` - basically removing `(` and `)`


When we are modifying a string we are getting a new string with changed characters (as desired). We can use method-chaining to perform several changes in one expression.

````python
df = pd.read_csv(url)

new_column_names = []
for column in df.columns:
    new_column_names.append(column.replace(' ', '_').replace('%', 'Percent').replace('(', '').replace(')', ''))

df.columns = new_column_names
df
````

[`.rename()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html) - we can use `rename` to change column name or names - we can do that for one or several columns.

A lot of functions that we have in Pandas have `inplace` argument, which is set to `False` and this means that the original DatFrame we are working on will not be changed and we will get a copy with changed data. If we use `inplace=True` then the original DataFrame will be changed.

````python
df.rename(columns={'Category': 'Item_category'})
````

````python
df.rename(columns={'Category': 'Item_category', 'Item': 'Product'})
````

# Data modifications

We have several dedicated methods to change the data within DataFrame or Series objects.
- [`.replace()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html) - we can think of this methods as a standard find/replace as from Word. If needed we can use regular expressions.
- [`.fillna()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html) - is used to fill/replace [missing values](https://pandas.pydata.org/docs/user_guide/missing_data.html)
- [`.apply()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) - allows us to applu any function (either buildin python function, pandas or our own: lambda or standard one defined by `def` keyword).

## `replace` - find and replace

This function can be applied to both whole DataFrame (all columns) or particular Series (particular column). Thats an another method that has `inplace` argument.

````python
df.replace('Breakfast', 'Petit-dejeuner').head(15)
````

````python
df.replace(23, -99)
````

````python
df.Saturated_Fat.replace(3.0, -100, inplace=True)
df
````

````python
df.Category.unique()
````

[`.sample()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html) - returns a random sample of rows from a DataFrame

````python
df.replace(['Desserts', 'Smoothies & Shakes'], 'Sweets').sample(15)
````

If I'm providing two lists, then I'm using a mapping between the value I'm looking for and value I'm replacing this value with. 

````python
df.replace(['Breakfast', 'Beverages'], ['Petit-dejeuner', 'Beuverages']).sample(15)
````

Another approach: I can provide a disctionary with columns (as key) and list of values I'm looking for (as values for this key) and replace those with something.

````python
df.replace({
    'Category': ['Breakfast', 'Beverages'],
    'Item': ['Egg McMuffin']
}, 'ABCD')
````

I can provide two dictionaries (both divided by columns) to indicate what I'm looking for and with what value it should be replaced, for a particular column.

````python
df.replace({
    'Category': ['Breakfast', 'Beverages'],
    'Item': ['Egg McMuffin']
}, {
    'Category': 'AAA',
    'Item': 'BBB'
})
````

We can use regular expressions (regex, regexp).

````python
# ex. McMuffin -> Muffin
# r'' -> raw, just leave as it is, don't substitute any characters inside starting with \
df.replace(r'Mc(\w+)', r'\1', regex=True)
````

## `apply`

Applies a function to each value from a DataFrame (row) or Series (value from the column).

```
Calories
300 -> times_10(300) -> 3000
250 -> times_10(250) -> 2500
370 -> times_10(370) -> 3700
```

````python
def times_10(number):
    return number * 10
````

````python
times_10(50)
````

When we are using `apply` with a function defined by `def` keyword we have to provide (as an argument) the name of the function. We can't use `()` after the, because this will execute the function. 

Correct: `df.Calories.apply(times_10)`

Incorrect `df.Calories.apply(times_10())`

````python
df.Calories, df.Calories.apply(times_10)
````

We can use `lambda` functions as well.

````python
def times_10(number):
    return number * 10
````

````python
df.Calories, df.Calories.apply(lambda number: number * 10)
````
