```python
def convert_height(s):
    feet, inches = s.split("-")
    return int(feet)*30.4 + int(inches)*2.54
```

```python
convert_height("6-1")
```

```python
df["Height"] = df.Height.apply(convert_height)
```

```python
# filtering
# getting playera taller than 200cm
df[ df.Height > 200 ].sort_values("Height")
```

```python
# filtering
# Height>200 and Salary>1mln$ and Age<25
df[ (df.Height>200) & (df.Salary>1_000_000) & (df["Age"]<25) ].sort_values(["Height","Salary","Age","Team"])
```

```python
# filtering with query() method
# Height>200 and Salary>1mln$ and Age<25
# df[ (df.Height>200) & (df.Salary>1_000_000) & (df["Age"]<25) ]
df.query(" Height>200 and Salary>1_000_000 and Age<25 ").sort_values(["Height","Salary","Age"])
```
