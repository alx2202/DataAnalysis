# Ex 1

Write a program that will display the results of multiplication of two numbers (1-10).

Sample output:
```
1*1=1
1*2=2
...
3*4=12
...
10*10=100
```

````python
# we need to use 2 for loops...
for number_1 in range(1, 11):
    for number_2 in range(1, 11):
        print(f'{number_1}*{number_2}={number_1*number_2}')
````

````python
a = 0
while a < 10:
    a += 1
    for b in range(11):
        print(f"The value is: {a}*{b}={a*b}")
````
